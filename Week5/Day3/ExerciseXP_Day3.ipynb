#### Exercise 7 

```python
# Create a line plot and scatter plot representing your favorite mathematical functions. 
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
```

```python
plt.scatter(x, y2, label='cos(x)')
```

#### Exercise 8

```python
# Plot the line plot and scatter plot from Exercise 7 on the same figure using subplots. 
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(x, y1, label='sin(x)')
ax2.scatter(x, y2, label='cos(x)')
```

#### Exercise 9 

```python
# Modify the plots from Exercise 8 to include a legend. 
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(x, y1, label='sin(x)')
ax1.legend(loc='upper left')
ax2.scatter(x, y2, label='cos(x)')
ax2.legend(loc='right')
```
